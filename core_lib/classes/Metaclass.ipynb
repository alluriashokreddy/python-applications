Metaclass_diagram.png

```
class Meta(type):
    def __init__(cls, name, bases, dct):
        print("meta init")
        super().__init__(name, bases, dct)

    def __new__(mcs, name, bases, dct):
        print("Creating")
        mcs.name = 'ashok'
        return super().__new__(mcs, name, bases, dct)
```

```
class Check(metaclass=Meta):
    value = 11

    def __init__(self, value):
        print('local init')
        self.value = value

    def display(self):
        print(self.value)

    @classmethod
    def display(cls):
        print(cls.value)

    @staticmethod
    def display():
        print("This is the value")
```
